## "Who has good wine, has a good friend." *italian proverb*

Wine is one of the oldest beverages in human history. It's the fermented juice of grapes, where the chemistry is carried by yeast that converts sugars into alcohol. However, each wine is the unique combination of soil, climate, and human craftsmanship.
The origins of winemaking trace back over 8,000 years to the South Caucasus region (modern Georgia), where the first evidence of grape fermentation was discovered. From there, viticulture spread through Mesopotamia, Egypt, and Greece, eventually becoming a central part of Roman culture. The Romans played a crucial role in expanding wine production across Europe, establishing vineyards in regions that remain famous today, such as Italy, Spain, Portugal, and France.
Throughout history, wine has held religious, social, and economic significance. In ancient Greece, it was a symbol of intellectual gatherings, while in medieval Europe, monasteries refined winemaking techniques that shaped modern viticulture. Today, wine is enjoyed worldwide, deeply embedded in culinary traditions and celebrations.


Wine is categorized mainly by the grape variety and production method. However, there are a few main types:

- Red Wine :wine_glass: ‚Äì Made from dark-skinned grapes, fermented with their skins to extract color and tannins.
- White Wine :clinking_glasses: ‚Äì Typically made from green or yellow grapes, fermented without skins for a lighter body.
- Ros√© Wine :cherry_blossom: ‚Äì Produced from red grapes with limited skin contact, resulting in a pink hue and delicate flavors.
- Sparkling Wine :champagne: ‚Äì Carbonated through natural or artificial methods, with Champagne being the most famous example.
- Dessert & Fortified Wines :tumbler_glass: ‚Äì Often sweet, with higher alcohol content due to added spirits (e.g., Port, Sherry, Madeira).

Wine production is a global industry, with an estimated 250‚Äì270 million hectoliters (hl) produced annually. The top-producing countries include:

- Italy (~50 million hl)
- France (~45 million hl)
- Spain (~35 million hl)

- Argentina üá¶üá∑ and Chile (~10‚Äì15 million hl each).

These countries dominate the market, however, high quality wines are produced also in smaller quantities and unexpected places.

### *Country choice: data selection*

The relevant fields in the dataframe for our purpose are: country, price, points, variety, winery, and taster_name. Therefore, I will:
- select only rows where all the values for those fields exist;
- clear the dataframe from duplicates.

Furthermore, I will focus on price and points ranges to select a suitable country. For a robust conclusion, I will consider only the countries with at least 1000 observations for price and points.

```python
# drop duplicated rows, but keep the first occurrence, and drop rows where any NaN occur in the fields of our interest
df_wine_country = df_wine.drop_duplicates(keep='first').dropna(subset=['country', 'price', 'points', 'variety', 'winery', 'taster_name'])

# group by country and count non-null observations (.size()) for price and points
country_counts = df_wine_country.groupby('country').size()

# select only the countries with min 1000 observations for price and points
thousand_countries = country_counts[country_counts >= 1000].index

# filter dataframe df_wine_country and keep only the countries with at least 1000 observations for price and points
df_wine_country = df_wine_country[df_wine_country['country'].isin(thousand_countries)]

# check which countries are left
df_wine_country.country.unique()
```

```python
# check the price range of the wine produced in these countries
df_wine_country.price.max(), df_wine_country.price.min()
```

```python
# and the rank range
df_wine_country.points.max(), df_wine_country.points.min()
```

```python
# the price range is pretty wide, let's check the occurrences per each price value
print(df_wine_country.price.value_counts().sort_index().to_string())
```

### *Country choice: data representation*

```python
# since the price range is very wide and the highest prices have very few occurrences (extreme values), the violinplot is not the best choice.
# I will use a boxplot with logarithmic scale for price distribution and a violinplot for the points distribution.
# p.s. the logarithmic scale should "compress" the outliers

# I need two subplots, one next to another, and each one with its own y scale (sharey=False)
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=False)

# I choose the palette
price_colors = sns.color_palette('Blues', 12)  
points_colors = sns.color_palette('Oranges', 12)  

# boxplot for log price distribution per country
sns.boxplot(data=df_wine_country, x='country', y='price', hue='country', palette=price_colors, ax=axes[0])

axes[0].set_yscale('log')
axes[0].set_title('Wine Price Distribution per Country')
axes[0].set_xlabel('Country')
axes[0].set_ylabel('log Price')
axes[0].tick_params(axis='x', rotation=45)

# violinplot for points distribution per country
sns.violinplot(data=df_wine_country, x='country', y='points', hue='country', palette=points_colors, ax=axes[1])
axes[1].set_title('Wine Rank Distribution per Country')
axes[1].set_xlabel('Country')
axes[1].set_ylabel('Points')
axes[1].tick_params(axis='x', rotation=45)

# for layout
plt.tight_layout()
plt.show()
```

Price distribution (boxplot): the y axis is on a logarithmic scale, hence the price differences are more noticeable. Most countries have a similar price distribution, however, some have more high-priced wines (outliers), such as France.

Rank distribution (violinplot): the wine score (points) is represented on the y axis and it's roughly between 80 and 100. From the width of the violins it's noticeable that most countries have a high concentration of wines scoring between 85 and 90 points, with a few countries reaching very high scores up to 100, such as Portugal and Italy.

To select the country that produces a good wine, available for a reasonable price, we can make the following considerations:
- the median price should be relatively low;
- the IQR (Interquartile Range) should be narrow, meaning that all the wines from that country will be around a reasonable price;
- the number of extreme outliers should be relatively low, because too many high-priced wines might indicate a premium market;
- the rank should be more towards the upper limit we identified earlier (85-90 points) to ensure a good quality.

Based on these considerations, **Portugal** seems to have a relatively lower median price while still offering quality wines.

### *Wineries producing the best Portuguese Red: data selection*

```python
# find the bin's limits
bins = np.histogram_bin_edges(df_selected['points'], bins=19)
print('Bin Edges:', bins)
```

```python
# the data of my interest fall in the bin 87.36842105 - 88.42105263. Now I check the wineries that produce the Portuguese Red wine in this rank range

# define the bin range of interest (whole numbers)
lower_bound, upper_bound = 87, 88

# filter wines within the bin
good_red_wines = df_selected[
    (df_selected['variety'] == 'Portuguese Red') & 
    (df_selected['points'] >= lower_bound) & (df_selected['points'] <= upper_bound)]

print(good_red_wines[['winery', 'points', 'province']].to_string())
```

### *Wineries producing the best Portuguese Red: data representation*

There are several wineries producing a high ranked, high quality Portuguese Red wine for a reasonable price. Let's check in which provinces are those wineries located and represent them on a map.

```python
# using Nominatim from the geopy.geocoders module, I can convert the provinces given in the dataset in coordinates, that I can further represent in a map

# create the object geolocator
geolocator = Nominatim(user_agent='portugal_wineries')

# list the provinces in the dataset with the Portuguese Red wine of our interest 
provinces = good_red_wines['province'].unique()

# create a dictionary to store the provinces coordinates
province_coords = {}

# get coordinates for each province
for province in provinces:  # iterate through the list provinces
    location = geolocator.geocode(province + ', Portugal')  # convert the provinces in coordinates only in Portugal
    if location:  # if the geolocator found the province...
        province_coords[province] = (location.latitude, location.longitude)  # the data (province name and coordinates) are stored in the dictionary province_coords
        print(f'{province}: {location.latitude}, {location.longitude}')
    else:  # if the geolocator didn't find the province...
        print(f'Could not find coordinates for {province}')
    time.sleep(1)  # to not have the timeout error, the Nominatim program is forced to wait one second between one request and the next

# convert the dictionary into a dataframe, so I can later represent the data in a plot
province_df = pd.DataFrame(province_coords.items(), columns=['province', 'coordinates'])  # convert the dictionary into a data frame where province and coordinates are two columns
province_df[['latitude', 'longitude']] = province_df['coordinates'].apply(pd.Series)  # the coordinates contain latitude and longitude, so I split them into two separated columns
province_df.drop(columns='coordinates', inplace=True)  # then I drop the original coordinates column

# merge the province_df dataframe with the dataframe collecting the high quality Portuguese Red wines
good_red_wines = good_red_wines.merge(province_df, on='province', how='left')
```

```python
# drop rows with NaN values in the good_red_wines dataframe
good_red_wines = good_red_wines.dropna(subset=['latitude', 'longitude'])

# create an interactive map using the folium package. The map is roughly centered in the Atlantic Ocean between Portugal and the Azores
portugal_map = folium.Map(location=[38.5, -18.0], zoom_start=6)

# add wineries to the map
for index, row in good_red_wines.iterrows():
    winery_name = row['winery']

    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=3,
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6,
        tooltip=f'{winery_name} - {row['province']}'
    ).add_to(portugal_map)

# save the map as an html file and display it
portugal_map.save('portugal_wineries.html')
portugal_map
```
